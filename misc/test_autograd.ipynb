```python
import scipy.stats
import autograd.numpy as np
from autograd import elementwise_grad as egrad 

from plot_functions import *

import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline
```

```python
def f(x):    
    mu = 0.0
    sigma = 1.0
    return (1.0/np.sqrt((2.0*np.pi*sigma**2))) * np.exp(-0.5*(x-mu)**2/sigma**2)
    #return scipy.stats.norm(loc=mu, scale=sigma).pdf(x)

def exact_gradient(x)    :
    mu = 0.0
    sigma = 1.0
    df = -(1.0/np.sqrt((2.0*np.pi*sigma**2))) * ((x - mu)/sigma**2) * np.exp(-0.5*(x-mu)**2/sigma**2)
    return df
    
auto_gradient_func = egrad(f)
```

```python
xvals = np.linspace(-3, 3, 100)
FONTSIZE = 22
plt.close('all')
fig, ax = plt.subplots(1, figsize=(10, 7))
ax.plot(xvals, f(xvals), lw=2, label='f')
ax.plot(xvals, exact_gradient(xvals), linestyle='-', label='exact gradient')
ax.plot(xvals, auto_gradient_func(xvals), marker='.', markersize=10, linewidth=0, label='auto grad')
ax.legend(loc='best', fontsize=FONTSIZE)
ax.set_title(r'$f = \mathcal{N}(0, 1)$', fontsize=FONTSIZE)
ax.set_xlabel('x', fontsize=FONTSIZE, labelpad=15)
ax.set_xlabel('f(x)', fontsize=FONTSIZE, labelpad=15)
adjustAxeProperties(ax, FONTSIZE, 0, FONTSIZE, 0)
```
