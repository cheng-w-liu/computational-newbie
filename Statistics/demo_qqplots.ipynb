Demo qqplot, the code snippet to generate the qqplot is motivated by the following blog post

http://scientificpythonsnippets.com/index.php/distributions/6-q-q-plot-in-python-to-test-if-data-is-normally-distributed

```python
import scipy.stats
import numpy as np
import pandas as pd
import seaborn as sbn
import matplotlib.pyplot as plt
import matplotlib
import statsmodels
from statsmodels.graphics.gofplots import qqplot
%matplotlib inline
matplotlib.style.use('ggplot')
```

```python
data = np.random.normal(loc=5, scale=2, size=200)
data.sort()

normal_dist = np.random.normal(size=len(data))
normal_dist.sort()

coeff = np.polyfit(normal_dist, data, 1)
polynomial = np.poly1d(coeff)

# Plot
FONTSIZE = 25
plt.close('all')
fig = plt.figure(figsize=(26, 8))
gs = matplotlib.gridspec.GridSpec(1, 3)

# 1st distribution
ax = plt.subplot(gs[0, 0])
ax = sbn.distplot(normal_dist, ax=ax, color='g')
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(FONTSIZE)
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(FONTSIZE)
ax.set_xlabel(r'$x$', fontsize=FONTSIZE, labelpad=15)
ax.set_ylabel(r'pdf($x$)', fontsize=FONTSIZE, labelpad=15)
ax.set_title('Normal distribution', fontsize=FONTSIZE)

# 2nd distribution
ax = plt.subplot(gs[0, 1])
ax = sbn.distplot(data, ax=ax, color='b')
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(FONTSIZE)
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(FONTSIZE)
ax.set_xlabel(r'$x$', fontsize=FONTSIZE, labelpad=15)
ax.set_ylabel(r'pdf($x$)', fontsize=FONTSIZE, labelpad=15)
ax.set_title('Normal distribution', fontsize=FONTSIZE)


# qqplot
ax = plt.subplot(gs[0, 2])
qq_plot, = ax.plot(normal_dist, data, marker='o', markersize=10, lw=0, color='r')
ax.plot(normal_dist, polynomial(normal_dist), 'k--')
ax.set_xlabel('quantiles of 1st dist.', fontsize=FONTSIZE, labelpad=15)
ax.set_ylabel('quantiles of 2nd dist.', fontsize=FONTSIZE, labelpad=15)
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(FONTSIZE)
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(FONTSIZE) 

plt.tight_layout(pad=0, w_pad=2.0, h_pad=1.0)
fig.suptitle('qqplot for normal versus normal', fontsize=FONTSIZE*1.1)
plt.subplots_adjust(top=0.85)
```

```python
beta_gen = scipy.stats.beta(10, 2)
data = beta_gen.rvs(size=200)
data.sort()

normal_dist = np.random.normal(size=len(data))
normal_dist.sort()

coeff = np.polyfit(normal_dist, data, 1)
polynomial = np.poly1d(coeff)

# Plot
FONTSIZE = 25
plt.close('all')
fig = plt.figure(figsize=(26, 8))
gs = matplotlib.gridspec.GridSpec(1, 3)

# 1st distribution
ax = plt.subplot(gs[0, 0])
ax = sbn.distplot(normal_dist, ax=ax, color='g')
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(FONTSIZE)
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(FONTSIZE)
ax.set_xlabel(r'$x$', fontsize=FONTSIZE, labelpad=15)
ax.set_ylabel(r'pdf($x$)', fontsize=FONTSIZE, labelpad=15)
ax.set_title('Normal distribution', fontsize=FONTSIZE)

# 2nd distribution
ax = plt.subplot(gs[0, 1])
ax = sbn.distplot(data, ax=ax, color='b')
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(FONTSIZE)
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(FONTSIZE)
ax.set_xlabel(r'$x$', fontsize=FONTSIZE, labelpad=15)
ax.set_ylabel(r'pdf($x$)', fontsize=FONTSIZE, labelpad=15)
ax.set_title('Beta distribution', fontsize=FONTSIZE)


# qqplot
ax = plt.subplot(gs[0, 2])
qq_plot, = ax.plot(normal_dist, data, marker='o', markersize=10, lw=0, color='r')
ax.plot(normal_dist, polynomial(normal_dist), 'k--')
ax.set_xlabel('quantiles of 1st dist.', fontsize=FONTSIZE, labelpad=15)
ax.set_ylabel('quantiles of 2nd dist.', fontsize=FONTSIZE, labelpad=15)
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(FONTSIZE)
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(FONTSIZE) 

plt.tight_layout(pad=0, w_pad=2.0, h_pad=1.0)
fig.suptitle('qqplot for Beta versus normal', fontsize=FONTSIZE*1.1)
plt.subplots_adjust(top=0.85)
```
