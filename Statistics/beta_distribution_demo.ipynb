Demonstrate Beta distribution $\text{Beta}(a, b)$. The first parameter $a$ corresponds to positive events, and the second parameter $b$ corresponds to negative events. The bigger the number $a+b$, the more certain the prediction and the more peaked the distribution.

```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
matplotlib.style.use('ggplot')
```

```python
a0 = 32
b0 = 109
mean0 = float(a0)/float(a0+b0)

a1 = a0+100
b1 = b0+200
mean1 = float(a1)/float(a1+b1)
```

```python
x = np.linspace(0, 1, 2000)
```

```python
beta0 = scipy.stats.beta(a=a0, b=b0)
y0 = beta0.pdf(x)
```

```python
beta1 = scipy.stats.beta(a=a1, b=b1)
y1 = beta1.pdf(x)
```

```python
XMAX = 1.05
YMAX = 1.1 * np.max([y0,y1])
FONTSIZE = 30
plt.close('all')
fig, ax = plt.subplots(1, figsize=(11, 8))
ax.set_xlabel('x', fontsize=FONTSIZE, labelpad=15)
ax.set_ylabel('Beta(x)', fontsize=FONTSIZE, labelpad=15)
ax.set_xlim(0, XMAX)
ax.set_ylim(0, YMAX)
ax.set_xticks(np.arange(0, XMAX, 0.1))
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(FONTSIZE)
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(FONTSIZE)
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_rotation(45)

p1, = ax.plot(x, y0, lw=2, color='r')
m1, = ax.plot([mean0, mean0], [0, beta0.pdf(mean0)], lw=1, linestyle='--', color='r')
p2, = ax.plot(x, y1, lw=2, color='b')
m2, = ax.plot([mean1, mean1], [0, beta1.pdf(mean1)], lw=1, linestyle='--', color='b')
plt.title('Demos of Beta distributions. \nDashed lines indicate the means.', fontsize=FONTSIZE*0.7)
l = plt.legend((p1, p2), 
               ('a={0:}, b={1:}'.format(a0, b0), 
                'a={0:}, b={1:}'.format(a1, b1)), 
                fontsize=FONTSIZE, loc='best')
```
