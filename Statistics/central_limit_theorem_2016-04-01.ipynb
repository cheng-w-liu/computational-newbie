Demonstration of Central Limit Theorem

Introduction:
(Motivated by Kerven Murphy's MLaPP, Chapter 2.)

Assume we have a random variable $X$, which has an unknown distribution from which we draw a sequence of independent and identically distributed (iid) values $x_i$, where $i=1 \dots N$.

Define the sample mean as $\bar{x} \equiv \frac{1}{N} \sum_i x_i$

Central Limit Theorem (CLT) states that the sample mean will be Gaussian-distributed as $N \rightarrow \infty$. (Notice that this is irrespective of the underlying distribution of $X$.)

Below we use two examples to demonstrate CLT, Beta distribution and uniform distribution. 

```python
import numpy as np
import scipy.stats
```

```python
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
matplotlib.style.use('ggplot')
```

```python
nRepeats = 10000
N_list = [1, 10, 100]
```

```python
# Beta distribution

a=1
b=5
beta_dist = scipy.stats.beta(a=a, b=b)

FONTSIZE = 20
plt.close('all')
fig = plt.figure(figsize=(12,12))
gs = matplotlib.gridspec.GridSpec(len(N_list), 1, width_ratios=[1, 1])

for i, N in enumerate(N_list):
    data = [np.mean(beta_dist.rvs(size=N)) for j in range(nRepeats)]
    counts, edges = np.histogram(data, bins=20)
    
    x = (edges[0:-1] + edges[1:])/2.0
    binWidth = edges[1] - edges[0]
    binHeight = 1.0/(binWidth * len(data))    
    y = binHeight * counts
    
    XMIN = 0.8 * np.min(x)
    XMAX = 1.1 * np.max(x)
    YMIN = 0.0
    YMAX = 1.1 * np.max(y)

    ax = plt.subplot(gs[i, 0])    
    p = ax.bar(x, y, width=binWidth)
    ax.set_xlim(XMIN, XMAX)
    ax.set_ylim(YMIN, YMAX)
    ax.set_xlabel(r'$x$', fontsize=FONTSIZE, labelpad=15)
    ax.set_ylabel(r'$p(x)$', fontsize=FONTSIZE, labelpad=15)
    for tick in ax.xaxis.get_major_ticks():
        tick.label.set_fontsize(FONTSIZE)
        tick.label.set_rotation(0)
    for tick in ax.yaxis.get_major_ticks():
        tick.label.set_fontsize(FONTSIZE)
        tick.label.set_rotation(0)        
    ax.set_title('N = {0:}'.format(N), fontsize=0.8*FONTSIZE)
plt.tight_layout(pad=0, w_pad=6.0, h_pad=3.0)   
```

```python
# uniform distribution 

FONTSIZE = 20
plt.close('all')
fig = plt.figure(figsize=(12,12))
gs = matplotlib.gridspec.GridSpec(len(N_list), 1, width_ratios=[1, 1])

for i, N in enumerate(N_list):
    data = [np.mean(np.random.random(size=N)) for j in range(nRepeats)]
    counts, edges = np.histogram(data, bins=20)
    
    x = (edges[0:-1] + edges[1:])/2.0
    binWidth = edges[1] - edges[0]
    binHeight = 1.0/(binWidth * len(data))    
    y = binHeight * counts
    
    XMIN = 0.8 * np.min(x)
    XMAX = 1.1 * np.max(x)
    YMIN = 0.0
    YMAX = 1.1 * np.max(y)

    ax = plt.subplot(gs[i, 0])    
    p = ax.bar(x, y, width=binWidth)
    ax.set_xlim(XMIN, XMAX)
    ax.set_ylim(YMIN, YMAX)
    ax.set_xlabel(r'$x$', fontsize=FONTSIZE, labelpad=15)
    ax.set_ylabel(r'$p(x)$', fontsize=FONTSIZE, labelpad=15)
    for tick in ax.xaxis.get_major_ticks():
        tick.label.set_fontsize(FONTSIZE)
        tick.label.set_rotation(0)
    for tick in ax.yaxis.get_major_ticks():
        tick.label.set_fontsize(FONTSIZE)
        tick.label.set_rotation(0)        
    ax.set_title('N = {0:}'.format(N), fontsize=0.8*FONTSIZE)
plt.tight_layout(pad=0, w_pad=6.0, h_pad=3.0)  
```

From the above examples, we can see that when $N=1$, the histogram shows the characteristic of the underlying distribution the r.v. is drawn from. As $N$ increases, the histogram becomes more and more like a Gaussian distribution.
